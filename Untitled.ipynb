# Análise dos dados da Copa do Mundo
Com base na página sobre a Copa do Mundo na Wikipedia vamos obter os dados de público pagante em cada final.
    
Link -> https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA

```python
# Importando as bibliotecas
import pandas as pd
import requests

# URl 
url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

# Requisitar a página e guardar a resposta
resposta = requests.get(url)

# Obter a resposta em formato HTML
table = pd.read_html(resposta.text)

# Olhando o resultado
table
```

```python
# Queremos apenas a tabela com pagantes por final.
# Essa é a 4ª tabela dos dados retornados 

df = table[3]
df.head()
```

```python
# Vamos analisar os tipos das colunas
df.info()
```

```python
df.columns = ['Ano', 'Sede', 'Nº Estadios/Cidades', 'Publico Total', 'Partidas', 'Publico Medio', 'Publico', 'Estadio', 'Jogo']
```

```python
df.info()
```

```python
df.head()
```

```python
# Precisamos converter a coluna Publico para int
df['Publico Total'] = df['Publico Total'].str.replace(" ", "").astype(int)
df.head()
```
