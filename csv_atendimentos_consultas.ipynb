```python
# Importando as bibliotecas
import pandas as pd
# Informar ao matplotlib sobre graficos inline
%matplotlib inline
```

```python
# Vamos importar apenas algumas colunas do arquivo
# 1 - Estabelecimento, 2 - Complexidade, 4 - Grupo, 5 - cod_subgrupo, 7 - Procedimento, 8 - Quantidade
# 9 - cod_carater_atendimento, 10 - Carater _atendimento

# Carregar arquivo csv
df = pd.read_csv('sia102017.csv', sep=',', encoding='cp1252', usecols=[1, 2, 4, 5, 7, 8, 9, 10])

# Trocar os nomes das colunas
df.columns = ['Hospital', 'Complexidade', 'Grupo', 'Cod. Supbgrupo', 'Procedimento', 'Quantidade'
              , 'Cod. Carater Atendimento', 'Carater Atendimento']

# Exibir as primeiras linhas do DataFrame
df.head(3)
```

```python
# Descrever as colunas númericas
df.describe()
```

```python
# Listar os hospitais presentes nos dados
df['Hospital'].unique()
```

```python
# Quantas consultas foram realizadas em cada hospital
df['Hospital'].value_counts()
```

```python
# Quantas consultas por subgrupo
df['Procedimento'].value_counts()
```

```python
# Plotar o gráfico por Procedimento
df['Procedimento'].head(200).value_counts().plot.bar()
```

```python
# Vamos plotar um gráfico de barras horizontal
df_formoterol['Hospital'].value_counts().plot.barh()
```

```python
# Melhorar o gráfico de barras horizontal
# Inverter a ordem & colocar o titulo
df_formoterol['Hospital'].value_counts(ascending=True).plot.barh(title='Quantidade de Formoterol por Hospital')
```
